**1) Ejercicio de herencia múltiple y polimorfismo:**

* Crea la clase BicicletaElectrica que herede de Vehiculos y de VElectricos, pero dando prioridad a VElectricos (ya que es un vehículo más electrico que normal)
* Crea la clase Quad que herede de Vehiculos y de VElectricos, pero dando prioridad a Vehiculos (ya que sólo usa la electricidad de modo puntual)

```python
class Vehiculos:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    def repostar():
        print("Este vehiculo tiene que repostar gasolina")


class VElectricos(Vehiculos):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo)
        self.autonomia = autonomia
        
    def repostar():
        print("Este vehiculo tiene que repostar electricidad")

def repostar(vehiculo):
    print("Este vehiculo tiene que repostar")

class BicicletaElectrica(VElectricos):
    def __init__(self, autonomia, marca, modelo):
        super().__init__(autonomia, marca, modelo)
        pass
    
class Quad(VElectricos):
    def __init__(self, marca, modelo, autonomia):
        super().__init__(marca, modelo, autonomia)
        pass
    

v_1 = Vehiculos('Ford', 'Mustang')
print (v_1.modelo)
velec_1 = VElectricos('Skoda', 'Kamiq', 100)
print(velec_1.marca)
print(velec_1.modelo)
print(velec_1.autonomia)
bicicleta_electrica_1 = BicicletaElectrica(150, 'Orbea', '550C')
quad_1 = Quad('Orbea', '550C', 150)

print(bicicleta_electrica_1.autonomia)
print(quad_1.autonomia)

repostar(bicicleta_electrica_1)
repostar(quad_1)
```

**2) Ejercicio de herencia**
* Crea la superclase Personal_Universitario con un único atributo que sea un diccionario que incluya: "id", "nombre" y "email"
* Crear la clases Oficina, Profesor y Alumno. Todas ellas heredaran de Personal_Universitario y heredaran sus datos
* En el caso de Oficina, debe añadir a su diccionario de datos el dato "Puesto"
* En el caso de Profesor, debe añadir a su diccionario de datos el dato "Especialización"
* En el caso de Alumno, debe añadir a su diccionario de datos el dato "CreditosAprobados" (integer)
* Crear métodos que sirvan para mostrar la información de cada clase
* Crear objetos de todas las clases
* Mostrar la información de esos objetos y comprobar que sale lo que tiene que salir

```python
# Completa el ejercicio aquí

class Personal_Universitario:
    def __init__(self, id, nombre, email):
        self.id = id
        self.nombre = nombre
        self.email = email
    def __str__(self):
        return """\
ID\t{}
NOMBRE\t\t{}
EMAIL\t\t{}""".format(self.id, self.nombre ,self.email)

class Oficina(Personal_Universitario):
    def __init__(self, id, nombre, email, puesto):
        super().__init__(id, nombre, email)
        self.puesto = puesto
        
    def __str__(self):
        return """\
ID: \t{}
NOMBRE\t\t{}
EMAIL\t\t{}
PUESTO\t\t{}""".format(self.id, self.nombre, self.email, self.puesto)
        
class Profesor(Personal_Universitario):
    def __init__(self, id, nombre, email, especializacion):
        super().__init__(id, nombre, email)
        self.especializacion = especializacion
        
    def __str__(self):
        return """\
ID\t{}
NOMBRE\t\t{}
EMAIL\t\t{}
ESPECIALIZACION\t\t{}""".format(self.id, self.nombre, self.email, self.especializacion)
        
class Alumno(Personal_Universitario):
    def __init__(self, id, nombre, email, creditos_aprobados):
        super().__init__(id, nombre, email)
        self.creditos_aprobados = int(creditos_aprobados)
        
    def __str__(self):
        return """\
ID\t{}
NOMBRE\t\t{}
EMAIL\t\t{}
CREDITOS APROBADOS\t{}""".format(self.id, self.nombre, self.email, self.creditos_aprobados)
        
personal_universitario1 = Personal_Universitario("0001", "Ismael", "ismael.garcia@uni.es")
print(personal_universitario1)

oficina_1 = Oficina("2000", "Clase 34", "admin@uni.es", "Departamento de Matematicas")
print(oficina_1)

alumno_1 = Alumno("0300", "Juan", "juan.gomez@uni.es", 48)
print(alumno_1)
```

```python
# Completa el ejercicio aquí

class Catalogo_ropa:
    def __init__(self, codigo, seccion, talla, precio):
        self.codigo = codigo
        self.seccion = seccion
        self.talla = talla
        self.precio = precio
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO \t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio)

class Ropa_abrigo(Catalogo_ropa):
    def __init__(self, codigo, seccion, talla, precio, material):
        super().__init__(codigo, seccion, talla, precio)
        self.material = material
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO \t\t{}
MATERIAL\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.material)
        
class Ropa_deporte(Catalogo_ropa):
    def __init__(self, codigo, seccion, talla, precio, tipo):
        super().__init__(codigo, seccion, talla, precio)
        self.tipo = tipo
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO\t\t{}
TIPO\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.tipo)
        
class Ropa_deporte_arriba(Ropa_deporte):
    def __init__(self, codigo, seccion, talla, precio, tipo, descripcion):
        super().__init__(codigo, seccion, talla, precio, tipo)
        self.descripcion = descripcion
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO \t\t{}
TIPO\t\t{}
DESCRIPCION\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.tipo, self.descripcion)
    
class Ropa_deporte_abajo(Ropa_deporte):
    def __init__(self, codigo, seccion, talla, precio, tipo, largo, descripcion):
        super().__init__(codigo, seccion, talla, precio, tipo)
        self.largo = largo
        self.descripcion = descripcion
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO \t\t{}
TIPO\t\t{}
LARGO\t\t{}
DESCRIPCION\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.tipo, self.largo, self.descripcion)
    
class Denim(Catalogo_ropa):
    def __init__(self, codigo, seccion, talla, precio, tipo, color):
        super().__init__(codigo, seccion, talla, precio)
        self.tipo = tipo
        self.color = color
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO \t\t{}
TIPO\t\t{}
COLOR\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.tipo, self.color)
    
class Zapatos(Catalogo_ropa):
    def __init__(self, codigo, seccion, talla, precio, tipo, color):
        super().__init__(codigo, seccion, talla, precio)
        self.tipo = tipo
        self.color = color
        
    def __str__(self):
        return """\
CODIGO\t{}
SECCION\t\t{}
TALLA\t\t{}
PRECIO\t\t{}
TIPO\t\t{}
COLOR\t\t{}""".format(self.codigo, self.seccion ,self.talla, self.precio, self.tipo, self.color)
    
abrigo1 = Ropa_abrigo("T388K3", "Ropa de abrigo", "XL", 79.99, "Plumón")
abrigo2 = Ropa_abrigo("T3WN53", "Ropa de abrigo", "XL", 129.99, "Cuero")
deporte1_1 = Ropa_deporte_arriba("A5N3LQ", "Deporte", "M", 29.99, "Sudadera", "Gorro y cremallera")
deporte1_2 = Ropa_deporte_arriba("A5JLE3", "Deporte", "S", 49.99, "Cortavientos", "Impermeable")
deporte2_1 = Ropa_deporte_abajo("F89JAB", "Deporte", "M", 29.99, "pantalon chandal", "Largo", "Tres bolsillos con cremallera")
deporte2_1 = Ropa_deporte_abajo("FR4GXK", "Deporte", "L", 19.99, "pantalon futbol", "Corto", "No tiene bolsillos")
jeans1 = Denim("F66BB", "Denim", "42", "34.99", "Jeans slim", "Lavabo oscuro")
jeans2 = Denim("DHA7A", "Denim", "44", "24.99", "Jeans carrot", "Lavabo medio")
chaqueta_vaquera1 = Denim("N7ABA9", "Denim", "L", 39.99, "Oversized", "Lavado claro")
chaqueta_vaquera2 = Denim("N733FX", "Denim", "XXL", 49.99, "Slim", "Negro")
zapatos1 = Zapatos("BSG573", "Zapatos", "45", 49.99, "Sneakers", "Blanco")
zapatos2 = Zapatos("BSF5G4", "Zapatos", "43", 69.99, "Running", "Azul")

productos = [abrigo1, abrigo2, deporte1_1, deporte1_2,deporte2_1, deporte2_1, jeans1, jeans2, chaqueta_vaquera1, chaqueta_vaquera2, zapatos1, zapatos2]

for i in productos:
    print(i, "\n")
```
